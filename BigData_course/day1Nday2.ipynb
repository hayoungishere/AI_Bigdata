```python
df = pd.DataFrame({"sample": [74.5,81.2,73.8,82.0,76.3,75.7,80.2,72.6,77.9,82.8]})

# 1-sample t-test 실행
t_result = sp.ttest_1samp(df, 76.7)

# t-test 실행결과에서 t검정통계량과 p-value를 받아온다.
t,p = t_result.statistic.round(3), t_result.pvalue.round(3)

print("1-Sample t-test")
print("t검정통계량 : %.3f"%t)
print("p-value : %.3f"%p)
```

## 2-sample t-test

    <가설검정 연습>
    H0 : 고객A와 고객B의 고객 만족도는 차이가 없다.
    H1 : 고객A와 고객B의 고객 만족도는 차이가 있다.

    p-value = 0.003
    따라서 p-value < 0.05(유의수준) 이므로, 귀무가설은 기각되고 대립가설을 채택된다.
    고객 A와 고개B의 고객 만족도 차이는 있다.

```python
df1 = pd.DataFrame({"sample":[6,5,5,4,6,7,6,4,5,6,4,5,5,6,4,8,6,5,6,7]})
df2 = pd.DataFrame({"sample":[7,5,7,8,7,8,8,5,7,6,5,5,6,6,5,7,9,7,7,8]})

# 2-sample t-test
t_result = sp.ttest_ind(df1,df2)

# t-test 실행결과에서 t 검정통계량과 p-value를 받아온다.
t,p = t_result.statistic.round(3), t_result.pvalue.round(3)

print("2-Sample t-test")
print("t검정통계량 : %.3f"%t)
print("p-value : %.5f"%p)
```

# ANOVA 
    3집단 이상일 때 평균의 차이를 알고 싶을 때 사용
    등분산성을 확인해서 평균을 보면되기 때문에....(?)
    군간변동이 클수록 f 검정통계량이 커진다. ==> 그에 따라 "차이가 있다"라고 볼 수 있다.

```python
df1 = pd.DataFrame({"A":[892,623,721,678,723,790,720,670,690,771]})
df2 = pd.DataFrame({"B":[721,821,910,678,723,790,711,790,745,891]})
df3 = pd.DataFrame({"C":[621,915,888,721,894,834,841,912,845,889]})

f_result= sp.f_oneway(df1, df2, df3)

f,p = f_result.statistic.round(3), f_result.pvalue.round(3)

print("One-way ANOVA")
print("F 검정통계량 : {}".format(f))
print("p-value : {}".format(p))

#Draw box-plot 
plot_data =[df1["A"], df2["B"],df3["C"]]
a = plt.boxplot(plot_data)
plt.show()
```

분산이 퍼져있는 것을 모으기는 어렵지만 분산이 모여있을 때는 평균을 이동시켜 서 다른 그룹과 차이가 없게 만들면 된다.

```python
from statsmodels.stats.proportion import proportions_ztest

count=15
nobs =100
value = .1

#Proportion test실행
stat, pval = proportions_ztest(count, nobs, value)

#Proportion test 결과 출력
print("1 Proportion test")
print("p 검정통계량 : {0:0.3f}".format(stat))
print("p-value : {0:0.3f}".format(pval))
```

```python
print("== 가설 수립 ==")
print("H0 : 고객만족도는 20%이다.")
print("H1 : 고객만족도는 20%가 아니다.\n")

print("== 가설 검정 ==")
value = .2
nobs =100
cnt =10

stat, pval = proportions_ztest(cnt, nobs, value)

#Proportion test 결과 출력
print("1 Proportion test")
print("p 검정통계량 : {0:0.3f}".format(stat))
print("p-value : {0:0.3f}\n".format(pval))


print("== Result == ")
print("p-value가 유의수준 0.05 보다 작기 때문에 귀무가설 '고객만족도는 20%이다.'는 기각된다.")
print("따라서, 품질개선의 결과로 고객 만족도의 변화가 생겼다고 할 수 있다.")
```

# 카이제곱 검정

```python
print("== 가설 수립 ==")
print("H0 : 근무조별로 제품을 생산하는 부하의 차이가 없다.")
print("H1 : 근무조별로 제품을 생산하는 부하의 차이가 존재한다.\n")

print("== 가설 검정 ==")
df = pd.DataFrame({"HSG":[270,260,236,234],"SS":[228,285,225,262],"SPA":[277,284,231,208]})

chi,pval, dof, expected = sp.chi2_contingency(df.T)

print("chi-square test")
print("chisq : %.3f"%chi)
print("p: %.3f"%pval)
print("degree of freedom : %d"%dof)
print("expected vlaue : \n {}\n".format(expected.round(3)))

print("== Result == ")
print("p-value가 유의수준 0.05 보다 작기 때문에 대립가설 '근무조별로 제품을 생산하는 부하의 차이가 존재한다.'는 채택된다.")
print("따라서, 근무조별로 제품을 생산하는 부하의 차이가 있다고 할 수 있다.")
```

```python
print("== 가설 수립 ==")
print("H0 : 이미지에 대해 차이가 없다.")
print("H1 : 이미지에 대해 차이가 존재한다.\n")

print("== 가설 검정 ==")
df = pd.DataFrame({"AS":[18,8,4,4,3,3],"price":[1,2,1,1,1,25], "spec":[8,14,3,2,3,8], "scalability":[7,5,4,3,1,10],
                  "Design":[10,5,9,2,1,2], "Secure":[9,9,5,7,1,1],"Functionality":[10,4,4,3,1,7]})

chi, pval, dof, expected = sp.chi2_contingency(df.T)

print("chi-square test")
print("chisq : %.3f"%chi)
print("p: %f"%pval)
print("degree of freedom : %d"%dof)
print("expected vlaue : \n {}\n".format(expected.round(3)))

print("== Result == ")
print("p-value가 유의수준 0.05 보다 작기 때문에 대립가설 '이미지에 대한 차이가 존재한다.'는 채택된다.")
print("따라서, 제페무별 이미지에 대하 차이가 존재한다.")
```
